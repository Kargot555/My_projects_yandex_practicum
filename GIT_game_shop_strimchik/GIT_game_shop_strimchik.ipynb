### Работа с пропусками

```python
# удалим непонятные игры без названия
data_games.dropna(subset=['name'],inplace=True)
```

Мы удалили пропуски в название игр и жанра соответственно, т.к.т не понятно, что это за игры да и их всего две

```python
data_games['year_of_release'].isna().sum()
```

```python
# используем лямбда функцию для замены пропущенных значений по годам,
# группируя по названию платформ и подсчитывая медиану
data_games['year_of_release'] = data_games.groupby('platform')['year_of_release'].apply(lambda x: x.fillna(x.median()))
```

```python
data_games['year_of_release'].isna().sum()
```

```python
# глянем строчку с ранее пропущенным годом
data_games[data_games['name'] == 'wwe Smackdown vs. Raw 2006']
```

Я заменил пропуски в годах медианой группируя по названию платформ. PS2, Sbox 360 и т.п. выходили в определенные года и игры должны быть где то там. Вряд ли на ps2 стоит ждать игру в 2020 году. Не для всех платформ это правда подойдёт, типа PC, но пропущенных значений около 1,5% и на результате не думаю, что сильно скажется

```python
data_games['critic_score'].isna().sum()
```

```python
# используем лямюда функцию для замены пропущенных значений 
# в оценке критиков, группируя по играм

# data_games['critic_score'] = data_games.groupby('name')['critic_score'].apply(lambda x: x.fillna(x.mean()))
```

```python
# используем лямюда функцию для замены пропущенных значений критиков группируя по оценке пользователей
# data_games['critic_score'] = data_games.groupby('user_score')['critic_score'].apply(lambda x: x.fillna(x.mean()))
```

Ничем не заменяю пропущенные значения, т.к. в дальнейшем мы будем проводить анализ по этим данным и мы не должны "рисовать" результат

```python
data_games['user_score'].isna().sum()
```

```python
# заменим tbd на NaN, т.к. он не мешает вычислительным операциям
data_games['user_score'] = pd.to_numeric(data_games['user_score'], errors='coerce')
data_games['user_score'].isna().sum()
```

Значение **tbd** - в играх обычно означает **to be Discussed**, идёт обсуждение, калибровка, рейтинга ещё нет. Мы заменили её на NaN, чтобы далее спокойно перевести в числовой формат и иметь возможность вести подсчёты с этим столбцом, просто ноль будет искажать результаты.

```python
# заполнил пропуски ещё одним рейтингом "нет рейтинга"
data_games['rating'] = data_games['rating'].fillna('no_rating')
data_games['rating'].isna().sum()
```

Этот столбец имеет строковый тип и заполнен различными рейтингами. Счёл логичным пропуски заполнить **"no_rating"**

### Работа с типами данных

```python
# столбец года к типу integer
data_games['year_of_release'] = data_games['year_of_release'].astype('float32').astype('Int64') 

data_games.info()
```

Мы привели столбцы к нужным типам данных для нас. Установили тип integer для *year_of_release*, причём сразу он не приводился к нему. Воспользовался советом сначала к float32, потом уже к integer. *user_score* после замены **tbd** на NaN приобрёл тип float64

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нам необходимо проверить гипотезу о различие или равенстве средний оценок на Xbox One и PC. Для этого мы берём актуальный нам срез данных. Формулируем нулевую гипотезу, что средние оценки на xbox и пк равны. Тогда альтернативная теория соответственно не равны

```python
# сделаем срез из актуального нам периода
data_games_xone = data[data['platform'] == 'XOne']
data_games_xone = data_games_xone['user_score']
data_games_xone.head()
```

```python
data_games_pc = data[data['platform'] == 'PC']
data_games_pc = data_games_pc['user_score']
data_games_pc.head()
```

```python
data_games_xone.mean()
```

```python
data_games_pc.mean()
```

```python
# проверяем нулевую гипотезу о равенстве средней оценки пользователе PC и XOne
# уровень статистической значимости
alpha = 0.05

results = st.ttest_ind(data_games_xone, data_games_pc, equal_var = False, nan_policy='omit') # nan_policy для игнорирования nan

print('p-значение:', results.pvalue)

if results.pvalue < alpha:  
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Мы сформулировали нулевую гипотезу, что средние оценки на пк и xbox one равны. После проверки актуального периода не получилось отвергнуть эту теорию, вполне оценки пользователей в среднем равны

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нам необходимо проверить, равны или отличаются средние оценки пользовательские оценки жанров Action и Sports. Создадим нулевыю теорию, что средние оценки равны. Тогда альтернативная, что оценки не равны.

```python
data_games_action = data[data['genre'] == 'Action']
data_games_action = data_games_action['user_score']
data_games_action.head()
```

```python
data_games_sports = data[data['genre'] == 'Sports']
data_games_sports = data_games_sports['user_score']
data_games_sports.head()
```

```python
# проверяем нулевую гипотезу о равенстве средней оценки пользователе жанров экшен и спорт
# уровень статистической значимости
alpha = 0.05

results = st.ttest_ind(data_games_action, data_games_sports, equal_var = False, nan_policy='omit') # nan_policy для игнорирования nan

print('p-значение:', results.pvalue)

if results.pvalue < alpha:  
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

После t-теста мы отвергаем нулевыю гипотезу, что оценки равны и можем предположить, что они отличны в среднем

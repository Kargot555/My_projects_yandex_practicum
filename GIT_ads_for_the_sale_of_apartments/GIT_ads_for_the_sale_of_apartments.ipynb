```python
# заменим высоты потолка на медиану, чтоб сильно большие значения не искажали
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

```python
# Замением пропущенные значения медианой, а то дальше мне функция не считает
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['floors_total'].isna().sum()
```

```python
# разницей общей и жилой заполнять не будем, появляются отрицательные значения, заменим медианой
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area'].median())
data['kitchen_area'].isna().sum()
```

```python
# заменим пропущенные в названия н.п. на "новосктройка"
data['locality_name'] = data['locality_name'].fillna('Новостройка')
data['locality_name'].isna().sum()
```

```python
# заменим пропущенные парки на 0
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['parks_around3000'].isna().sum()
```

```python
# заменим пропущенные водоёмы на 0
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
data['ponds_around3000'].isna().sum()
```

```python
# в задаче нам поставили вопрос оценки стоимости удаления от центра в Питере, поэтому и заменим на медиану для Питера
data[data['locality_name'] == 'Санкт-Петербург']['cityCenters_nearest'].median()
```

- В название населённых пунктов ввели новое название "Новостройка" преполагая, что это они и названия ещё не получили
- Для парков и озёр в радиусе 3км пропущенные значения заменили на ноль, предполагая, что они там и должны быть, автоматическая программа их не нашла
- В площадях помещений поставили медиану, т.к. она менее подвержена выбросам. Попробовали дополнительные операции, например вычесть из общей кухни и получить жилую - на результат это не повлияло
- В столбце времени продажи пропущенные значения скорее всего показывают то, что квартира ещё не продана. Удалять строки с ними нельзя, потеряем данные в других столбцах, заменять на медиану не станем, это сильно повлияет на график

Проверяем добавленные столбцы

```python
# созданим функцию для создания категорий для этажей
def floors_total_floor(row):
    floors_total = row['floors_total']
    floor = row['floor']
    
    if floors_total == floor:
        return 'Последний_этаж'
    if floor == 1:
        if floors_total > floor:
            return 'Первый_этаж'
    return 'Другой'

# создадим новый столбец, пременив функцию к столбцу "этаж"    
data['floor_type'] = data.apply(floors_total_floor, axis=1)
data.head()       
```
